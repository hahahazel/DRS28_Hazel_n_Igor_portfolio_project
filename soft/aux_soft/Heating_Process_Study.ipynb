```python
path = '../Measurements_n_Tests/GGSv2_heating_measurements/'
file_name = 'GGSV2_Ambient_log_20211120.txt'
path_file = path + file_name
# data_f = np.loadtxt(path_file,  delimiter=',')
data_f_20Nov = np.loadtxt(path_file, delimiter=',')

data2plot = data_f_20Nov
plt.figure(figsize=(15, 10))
plt.plot(data2plot[:,0]-data2plot[0,0], data2plot[:,1], '-', label="B102 NO2", color="#1f77b4")
plt.plot(data2plot[:,0]-data2plot[0,0], data2plot[:,2], '-', label="B302 C2H5OH", color="#ff7f0e")
plt.plot(data2plot[:,0]-data2plot[0,0], data2plot[:,3], '-', label="B502 VOC", color="#2ca02c")
plt.plot(data2plot[:,0]-data2plot[0,0], data2plot[:,4], '-', label="B702 CO", color="#d62728")
plt.legend(loc="upper right")
plt.title("Groove Gas Sensor v2 readings during Heating process"+" File: "+file_name)
plt.xlabel("time [sec]")
plt.ylabel("Readings [ppm]")
plt.ylim([0,800])
plt.grid(True)
# plt.savefig("20211120_GGSv2.png")
plt.show()
plt.close()
```
