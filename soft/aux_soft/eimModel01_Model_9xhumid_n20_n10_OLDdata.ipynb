```python
# import os
# import sys, getopt
# import signal
# import time
from edge_impulse_linux.runner import ImpulseRunner
# import io
import matplotlib.pyplot as plt
```

```python
pathN = "../Measurements_n_Tests/DataSet_Run1/ai-sense-1st-run-export_all_manual_split/"
fileN = "proj6test_manualsplit_9xhumid_n20_n10-linux-x86_64-v1.eim"
modelfile = pathN + fileN

runner = ImpulseRunner(modelfile)
```

```python
model_info = runner.init()
print('Loaded runner for "' + 
      model_info['project']['owner'] + 
      ' / ' + 
      model_info['project']['name'] + 
      '"')
```

```python
features = [842.0, 548.0, 736.0, 340.0] #carrot
res = runner.classify(features)
print("classification:")
print(res["result"])
print("timing:")
print(res["timing"])
```
